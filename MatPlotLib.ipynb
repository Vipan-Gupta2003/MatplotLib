Object Oriented Way

Matplotlib allows the aspect ratio,DPI and figure size to be specified when the figure object is created. you can use thefigsize and dpi keyword argument

figsize is a tuple of the width and height of the figure in inches

dpi is the dots-per-inches (pixel per inch)

```python
fig = plt.figure()
```

```python
fig.add_axes([0.1,0.1,0.8,0.8]) # value will be between 0 and 1
fig
# first argument : shift from left or toward x -axis
# second argument  : shift from bottom or toward y axis
# third argument  : width
# fourth argument : height
```

```python
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
fig
```

```python
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.1,1,1])
axis2.plot(x,y)
fig
```

```python
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.1,0.1,0.3,0.3])
axis2.plot(x,y)
fig
```

```python
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.1,0.3,0.3])
axis2.plot(x,y)
fig
```

```python
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)
fig
```

```python
fig = plt.figure(figsize = (6,8))
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)
# fig
```

```python
fig = plt.figure(figsize = (6,8),dpi = 500)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize = (6,8),dpi =100)
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.3,0.3,0.3])
axis2.plot(x,y)
```
