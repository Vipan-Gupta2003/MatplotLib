**Now here come some basic questions**

1. which team won the most matches ?


```python
df.head(2)
```

```python
match_win = df['match_winner'].value_counts()
```

```python
sns.barplot(y=match_win.index , x= match_win.values,palette='viridis')
plt.title('Most Matches Won by Teams')
```

2. Toss Decision Trend

```python
sns.countplot(x=df['toss_decision'],palette='rainbow')
plt.title('Toss Decision Trend')
```

3. Toss Winner vs Match Winner

```python
count = df[df['match_winner']==df['toss_winner']]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

4. how do team win ? (run vs wicket)

```python
sns.countplot(x=df['won_by'])
```

5. Most "Player of the match" award

```python
sns.countplot( df['player_of_the_match'],palette = 'rainbow')
```

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values, y =count.index, palette='mako')
plt.title('Most Player of the Match Awards')
```

6. 2 top scorers

```python
count = df['top_scorer'].value_counts()
count
```

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind = 'bar')
```

7. 10 best bowling figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
```

```python
df
```

```python
df['highest_wickets'] = df['highest_wickets'].astype(int)
```

```python
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind='bar')
```

8. Most Matches Played by Venue

```python
venue_count = df['venue'].value_counts()
venue_count
```

```python

sns.barplot(y = venue_count.index,x = venue_count.values,palette='rainbow')
```

9.  Who won the highest margin by runs?

```python
df[df['won_by'] == 'Runs'].sort_values(by = 'margin',ascending=False).head(1)[['match_winner','margin']]
```

10. Which player had the highest individual score?

```python
df[df['highscore'] == df['highscore'].max()][['top_scorer','highscore']]
```

11. Which bowler had the best bowling figures?
