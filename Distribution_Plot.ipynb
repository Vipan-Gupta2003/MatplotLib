We use distribution plots to visualize the distribution of quantitative data/ continuous data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

Dis Plot

```python
sns.displot(df['total_bill'])
```

```python
sns.displot(df['total_bill'], kde =True)
```

```python
sns.displot(df['total_bill'], kde =True,bins = 20)
# kde = kernel density estimator
```

```python
sns.histplot(df['total_bill'], kde =True,bins = 20)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], kde =True,bins = 20)
plt.subplot(1,2,2)
sns.histplot(df['tip'], kde =True,bins = 20)
```

Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='sex',palette='rainbow')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

Rug Plot

```python
sns.rugplot(df)
```

```python
sns.rugplot(df['total_bill'])
```
